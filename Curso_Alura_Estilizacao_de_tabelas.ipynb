# Missão 4: Time de logística

Novamente o time de logística deseja repensar sua estrutura e organização de fornecimento de produtos e para isso, foi pensada em uma visualização que permita identificar padrão de **quantidade de produtos solicitados por mês por departamento**.

```python
df_meses = df.copy()
df_meses = df_meses.sort_values('data_pedido')
df_meses
```

```python
df_meses['meses'] = df_meses['data_pedido'].dt.strftime('%Y - %b')
df_meses = df_meses.reset_index(drop=True)
df_meses
```

```python
vendas_mensais = df_meses.pivot_table(index= 'departamento', columns= 'meses', values= 'quantidade',
                                      aggfunc= 'sum', sort= False)
vendas_mensais
```

```python
estilo_mensal = vendas_mensais.style
estilo_mensal
```

```python
estilo_mensal.set_sticky(axis= 'index')
```

```python
colunas = {
    'selector': '.col_heading',
    'props': 'font-weight: normal; font-family: Arial'
}
tabelas = {
    'selector': 'td,th',
    'props': 'text-align: left;'
}
index = {
    'selector': '.index_name',
    'props': 'font-weight: bold; font-family: Arial; text-align: right;'
}
estilo_mensal.set_table_styles([colunas,tabelas,index],overwrite=False)
```

Desafio: Nosso desafio agora é fornecer uma visualização que permita mostrar o tempo médio de entrega de pedidos durante os meses dos anos de 2013 a 2015 por cada país. Isso vai permitir analisar o desempenho da parte de transportes da empresa durante os anos e traçar um planejamento de melhorias.

```python
df_desafio.head()
```

```python
df_tempo_entrega = df_desafio.copy()
df_tempo_entrega = df_tempo_entrega.sort_values('data_pedido')
df_tempo_entrega = df_tempo_entrega.reset_index(drop=True)
df_tempo_entrega
```

```python
df_tempo_entrega = df_tempo_entrega[df_tempo_entrega['data_pedido'] >= '2013-01-01']
df_tempo_entrega['meses'] = df_tempo_entrega['data_pedido'].dt.strftime('%Y - %b')
tempo_entrega = (df_tempo_entrega['data_chegada'] - df_tempo_entrega['data_pedido']).dt.days
df_tempo_entrega['Tempo de Entrega'] = tempo_entrega
tempo = df_tempo_entrega.pivot_table(index= 'pais', columns= 'meses', values= 'Tempo de Entrega',
                                     aggfunc='mean', sort=False)
tempo
```

```python
estilo_tempo = tempo.style
estilo_tempo = estilo_tempo.format('{:.2f} dias')
estilo_tempo.set_sticky(axis='index')
```

```python
colunas = {
    'selector': '.col_heading',
    'props': 'font-weight: normal; font-family: Calibri'
}
tabelas = {
    'selector': 'td,th',
    'props': 'text-align: center;'
}
index = {
    'selector': '.index_name',
    'props': 'font-weight: bold; font-family: Arial; text-align: right;'
}
estilo_tempo.set_table_styles([colunas,tabelas,index],overwrite=False)
```

# Missão 5: Relatório de performance

O relatório de performance é um documento muito importante para a empresa. Nele, todos que trabalham na loja podem entender como anda a empresa. Por conta disso, tabelas bem construídas são essenciais para mostrar os dados da melhor forma.

Portanto, precisamos construir uma visualização que permita entender a **relação entre tipos de clientes e modo de envio de produtos de acordo com as vendas**. Com isso, será possível destacar no relatório qual o modo de envio e quais tipos de modo de envio mais geraram ganho em vendas.

```python
df_cliente = df.pivot_table(index= 'segmento_cliente', columns= 'modo_envio', values= 'vendas',
                            aggfunc= 'sum')
df_cliente
```

```python
df_cliente['Total'] = df_cliente.sum(axis=1)
df_cliente.loc['Total']= df_cliente.sum()
df_cliente
```

```python
compra_cliente = df_cliente.style.format('{:,.2f}')
compra_cliente
```

```python
tabela = {
    'selector': 'td, th:not(.index_name)',
    'props': 'font-weight: normal; font-family: Arial; text-align: center; background-color: white;'
}
index = {
    'selector': '.index_name',
    'props': 'font-weight: normal; text-align: right; font-style: italic; color: #696969;'
}
compra_cliente.set_table_styles([tabela, index])
```

```python
compra_cliente.set_table_styles({
    'Total': [{
        'selector':'th',
        'props': 'border-top: 1px solid #181818'
    },
    {
        'selector':'td',
        'props': 'border-top: 1px solid #181818'
    }],
    'B2B': [{
        'selector':'th',
        'props': 'border-top: 1px solid #181818'
    },
    {
        'selector':'td',
        'props': 'border-top: 1px solid #181818'
    }]
}, overwrite= False, axis=1)
```

```python
compra_cliente.set_table_styles({
    'Total': [{
        'selector': '.true',
        'props': 'background-color: #D8D8D8'
    }]
}, overwrite= False, axis=0)

cores_coluna = pd.DataFrame(['false', 'true', 'false'], index= df_cliente['Total'].index,
                            columns= ['Total'])
compra_cliente.set_td_classes(cores_coluna)
```

```python
compra_cliente.set_table_styles({
    'Total': [{
        'selector': '.true',
        'props': 'background-color: #D8D8D8'
    }]
}, overwrite= False, axis=1)

cores_linha = pd.DataFrame([['false','false', 'true', 'false', 'false']], columns = df_cliente.columns,
                            index= ['Total'])

compra_cliente.set_td_classes(cores_linha)
```

Desafio: envolve uma visualização que permita separar os gastos de envio de cada produto (custo total) por tipo de cliente, de modo que seja possível entender o tipo de produto que mais gerou custos de envio em cada tipo de cliente. Além disso, é interessante também mostrar a distribuição desses custos por país enviado, permitindo entender também os gastos por país.

```python
df_nome = df_desafio.rename(columns={'pais': 'Custo por País'})
df_pais_custo = df_nome.pivot_table(index=['tipo_cliente', 'categoria'], columns= 'Custo por País', values = 'custo_produto', aggfunc= 'sum')
df_pais_custo
```

```python
df_pais_custo.rename_axis(index={'tipo_cliente': 'Tipo de cliente', 'categoria': 'Categoria do produto'}, inplace = True)
df_pais_custo['Total'] = df_pais_custo.sum(axis=1)
df_pais_custo
```

```python
estilo_pais = df_pais_custo.style.format('${:,.2f}')
estilo_pais
```

```python
tabela = {
    'selector': 'td, th, th:not(.index_name)',
    'props': 'font-weight: normal; font-family: Calibri; text-align: center; background-color: white;'
}
index = {
    'selector': '.index_name',
    'props': 'font-weight: bold; font-family: Calibri; text-align: right;'
}
estilo_pais.set_table_styles([tabela, index])
estilo_pais.set_table_styles({
    ('B2B', 'BlueRay'): [{'selector': 'th', 'props': 'border-top: 1px solid #181818'},
              {'selector': 'td', 'props': 'border-top: 1px solid #181818'}],
    ('B2B', 'Livro'): [{'selector': 'th', 'props': 'font-weight: bold'}],
    ('B2C', 'Livro'): [{'selector': 'th', 'props': 'font-weight: bold'}],
}, overwrite=False, axis=1)

estilo_pais.set_table_styles({
    'Total': [{
        'selector': '.true',
        'props': 'font-weight: bold; color: blue; background-color: #D8D8D8'
    }]
},overwrite=False,axis=0)

cores_coluna = pd.DataFrame(['false','false','false','true','false','false','false','true'],index= df_pais_custo['Total'].index,
                            columns = ['Total'])

estilo_pais.set_td_classes(cores_coluna)
```
